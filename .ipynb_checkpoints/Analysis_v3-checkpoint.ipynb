```python
import networkx as nx
from matplotlib import colors
from networkx.drawing.nx_agraph import graphviz_layout

def make_gate_pair_df(subset):
    s = subset.groupby(['ExitGate','EntryGate'])['car-id'].count().reset_index()
    s.rename(columns={"car-id": "weight"},inplace=True)
    #display(s)
    return s

def make_gate_pair_dict(subset):
    ## Prepare the data
    gate_pair_dict = {}
    for index, series in subset.iterrows():
        from_gate, to_gate = series['ExitGate'],series['EntryGate']
        gate_pair_dict[(from_gate,to_gate)] = gate_pair_dict.get((from_gate,to_gate), 0) + 1
    return gate_pair_dict

def add_edge(G, metadata,weight):
    ## Add the metadata to each edge
    G.add_edge(metadata['from_gate'], metadata['to_gate'], weight=weight, **metadata)
        
def make_graph_G(subset,Digraph=False):
    with Timer("make_gate_pair_dict"):
        gate_pair_df = make_gate_pair_df(subset)
    
    
    with Timer("Build Graph"):
        if Digraph:
            G = nx.DiGraph()
        else:
            G = nx.Graph()
        # G.add_edge(1,2)
        for index, series in gate_pair_df.iterrows():
            from_gate, to_gate, weight = series['ExitGate'],series['EntryGate'], series['weight']
            metadata = {
                "from_gate": from_gate,
                "to_gate": to_gate,
            }
            add_edge(G,metadata, weight)
    return G,gate_pair_df 

def Draw_Graph_Util_With_Centrality_and_Cluster(G, subset, gate_feq_dict, custompos,figsize = (10,10),edge_scale=0.5215, node_scale=10000.0, centality_type="Degree"):
    # plot it
    random.seed(43)
    values = list(colors.cnames.keys())
    random.shuffle(values)
    # BASED ON wight
    edge_width = [edge_scale * G[u][v]['weight'] for u, v in G.edges()] 
    
    ## with deg_centrality
    if centality_type == CentralityTypes.Degree:
        centrality = nx.degree_centrality(G)
    elif centality_type == CentralityTypes.Betweenness:
        centrality = nx.betweenness_centrality(G)
    elif centality_type == CentralityTypes.Closeness:
        centrality = nx.closeness_centrality(G)
        
    node_size = [ node_scale *centrality.get(v,0) for v in G] 
    
    #COMPUTE THE COMMUNITY - CLSUTER
    parts = community.best_partition(G)
    c_values = [parts.get(node) for node in G.nodes()]
    node_color = [ values[n] for n in c_values]
    df = DF(parts)
    df['node_color'] = node_color
    node_color_dict = df.node_color.to_dict()
    node_color = [node_color_dict.get(v, "red") for v in G] 
    
    fig= plt.figure(figsize=figsize)
    a1 = plt.subplot2grid((3,3),(0,0), rowspan = 2, colspan = 2)
    a1.axis('off')
    a2 = plt.subplot2grid((3,3),(0,2), rowspan = 2)
    #network graph
    nx.draw_networkx(G, pos=custompos,
                     node_size = node_size,
                     node_color = node_color, alpha = .8, 
                     with_labels = True, width = edge_width, 
                     edge_color ='.5', cmap = plt.cm.Blues, ax=a1)
    #plt.axis('off')
    #bar chart
    gate_feq      = pd.Series(subset.EntryGate.value_counts().to_dict())
    plot = sns.barplot(gate_feq.values,gate_feq.index,color='skyblue', ax=a2)
    _  = plot.set_xticklabels(plot.get_xticklabels(), rotation=90)
    
    plt.tight_layout();
    return fig
    
def filter_date(df,Date,figsize=None, edge_scale=0.5215, node_scale=10000.0):
    ## plot the graph h,ere...
    subset = df.loc[df.Date == Date]
    G,gate_feq_dict = make_graph_G(subset)
    return Draw_Graph_Util_With_Centrality_and_Cluster(G, subset, gate_feq_dict, custompos,figsize,edge_scale=edge_scale, node_scale=node_scale)
    #return subset.shape
```

```python
import time
class Timer():
    def __init__(self,name=""):
        self.name = name
    def __enter__(self):
        self.s = time.time()
    def __exit__(self,a,b,c):
        self.e = time.time()
        print(f"Module:{self.name} \t Took: {round(self.e-self.s, 2)} seconds")
                       
```

```python
class CentralityTypes():
    Degree = "Degree" 
    Betweenness = "Betweenness"
    Closeness = "Closeness"
```

## Database UTIL

```python
from pymongo import MongoClient
from datetime import datetime
import pandas as pd

class MongoDb():
    def __init__(self, uri="mongodb+srv://admin:admin@cluster0-sjuby.mongodb.net/?retryWrites=true&w=majority"):
        self.uri = uri
        self.client = MongoClient(self.uri)
        
    def collection(self,name):
        return self.client['mydb'][name]

    def insert_one(self,name,doc):
        return self.client['mydb'][name].insert_one(doc)
    
    def delete(self,name):
        return self.client['mydb'][name].delete_many({})
    
    def show(self,name, limit=5):
        return pd.DataFrame(self.client['mydb'][name].find({}))
    
    def helper_get_all_offers(self,name):
        return pd.DataFrame(self.client['mydb'][name].find({}))
    
    def helper_get_offers_by_gates(self, name, gates=[]):
        Q = { "gate": { "$in": gates } }
        return pd.DataFrame(self.client['mydb'][name].find(Q))
        
        
    def helper_get_offers_by_gate(self, name, gate):
        return pd.DataFrame(self.client['mydb'][name].find({
            "gate": gate
        }))
    
    def helper_get_offers_by_vendor(self,name,vendor):
        return pd.DataFrame(self.client['mydb'][name].find({
            "vendor_name": vendor
        }))
    
    def helper_get_offers_by_gate_and_vendor(self,name,gate,vendor):
        return pd.DataFrame(self.client['mydb'][name].find({
            "gate": gate,
            "vendor_name": vendor
        }))
    
db = MongoDb()
```

# Tab Generation Utils

```python
def generate_tabs(names:list, fns:list):
    outs = []
    for tab_name, func in zip(names,fns):
        out = widgets.Output()
        with out:
            clear_output()
            func()
        outs.append(out)
    
    tab = widgets.Tab(children = outs)
    for i,name in enumerate(names):
        tab.set_title(i, name)
    display(tab)
    store = {"tab" : tab, "outs": outs }
    return store
```

## Vendor Dashboard

```python
def dashboard_add_offers():
    print("Add Offers page")
    valid_gates = list(clean_data.EntryGate.unique())
    gate_dropdown = widgets.Dropdown(options=valid_gates, value=valid_gates[0], description='Gate:')
    
    valid_vendors = ['vendor1', 'vendor2', 'vendor3', 'vendon4']
    vendor_dropdown = widgets.Dropdown(options=valid_vendors, value=valid_vendors[0], description='Vendor:')
    
    text_box = widgets.Text(description='Offer name')
    savebtn = widgets.Button(description="Publish Offer")

    out = widgets.Output()
    GATE_SELECTED, VENDOR_SELECTED = valid_gates[0], valid_vendors[0]

    def add_offer(v):
        with out:
            clear_output()
            if text_box.value != "":
                doc = {
                    "name": text_box.value,
                    "ts": datetime.now(),
                    "vendor_name": VENDOR_SELECTED,
                    "gate": GATE_SELECTED
                }
                db.insert_one( "offers", doc)
                print("Offer Published")
                
            
    
    def on_change_gate_dpdown(change):
        nonlocal GATE_SELECTED
        with out:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                GATE_SELECTED = change['new']
    
    def on_change_vendor_dpdown(change):
        nonlocal VENDOR_SELECTED
        with out:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                VENDOR_SELECTED = change['new']
                
    #register a change listener
    gate_dropdown.observe(on_change_gate_dpdown)
    vendor_dropdown.observe(on_change_vendor_dpdown)
    
    savebtn.on_click(add_offer)
    
    # render the box
    controls = widgets.HBox([vendor_dropdown, gate_dropdown, text_box, savebtn])
    W = widgets.VBox([controls, out])
    display(W)
    
def dashboard_list_offers():
    print("List Offers page")
    valid_gates = ["ALL"] + list(clean_data.EntryGate.unique())
    gate_dropdown = widgets.Dropdown(options=valid_gates, value=valid_gates[0], description='Gate:')
    out = widgets.Output()
    
    valid_vendors = ["ALL"] + ['vendor1', 'vendor2', 'vendor3', 'vendon4']
    vendor_dropdown = widgets.Dropdown(options=valid_vendors, value=valid_vendors[0], description='Vendor:')
    
    GATE_SELECTED, VENDOR_SELECTED = valid_gates[0], valid_vendors[0]

    def render():
        
        clear_output()
        
        #Filter offer data from DB
        if VENDOR_SELECTED == "ALL" and GATE_SELECTED == "ALL":
            dbdata = db.helper_get_all_offers("offers")
        elif VENDOR_SELECTED == "ALL":
            dbdata = db.helper_get_offers_by_gate("offers",GATE_SELECTED)
        elif GATE_SELECTED == "ALL":
            dbdata = db.helper_get_offers_by_vendor("offers",VENDOR_SELECTED)
        else:
            dbdata = db.helper_get_offers_by_gate_and_vendor("offers",GATE_SELECTED, VENDOR_SELECTED)
        print(f"Offers Found : {len(dbdata)}")
        display(dbdata)
        
        if len(dbdata) > 0:
            dbdata.rename(columns={"name": "offer"},inplace=True)
            dbdata = dbdata[['offer', 'ts', 'vendor_name', 'gate']].reset_index()
            sns.countplot(x="gate", data=dbdata, hue="vendor_name")
            plt.show()
        
          
    def on_change_vendor_dpdown(change):
        nonlocal VENDOR_SELECTED
        with out:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                VENDOR_SELECTED = change['new']
                render()
                  
    def on_change_gate_dpdown(change):
        nonlocal GATE_SELECTED
        with out:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                GATE_SELECTED = change['new']
                render()
    #register a change listener
    gate_dropdown.observe(on_change_gate_dpdown)
    vendor_dropdown.observe(on_change_vendor_dpdown)

    # render the box
    controls = widgets.HBox([vendor_dropdown, gate_dropdown])
    W = widgets.VBox([controls, out])
    display(W)
    
## Setup the dashboard
print("Vendor Dashboard")
store = generate_tabs(names=["List Offers","Add Offers"], fns= [dashboard_list_offers, dashboard_add_offers])
```

```python
## Util to forecast using Prophet
def forecast_util(subset, duration=10, display_forecast=True):
    """
    duration: no of days to forecast
    """
    ##get data to fit
    tmp = subset.sort_values(['Date_dt'])
    tmp = tmp.groupby("Date_dt")['Date_dt'].count()
    #tmp.plot(kind="line")
    tmp = pd.DataFrame(tmp)
    tmp.columns = ['y']
    tmp = tmp.reset_index()
    tmp.columns = ['ds','y']
    
    #fit and predict (plot as well)
    model = Prophet()
    model.fit(tmp)
    future = model.make_future_dataframe(periods=duration)
    forecast = model.predict(future)
    fig1 = model.plot(forecast)
    plt.show()
    if display_forecast:
        display(forecast.iloc[-(duration+3):][['ds','yhat','yhat_lower','yhat_upper']])
    return model, forecast, fig
```

```python
import calplotcustom as cp

def plot_heatmap(subset,year=2015,ax=None):
    if ax != None:
        ax.clear()
    def unique_carid_count(x):
        return len(x.unique())
    unique_car_counts= subset.groupby( [subset.Date_dt.dt.year, subset.Date_dt.dt.month, subset.Date_dt.dt.day] )['car-id'].apply( unique_carid_count)

    tmp = pd.DataFrame(unique_car_counts.items())
    tmp.columns=['date','value']
    tmp.date = pd.to_datetime(tmp.date.apply(lambda x: "/".join(map(str, (x)))))
    tmp['year'] = tmp.date.dt.year
    tmp['weekno'] = tmp.date.dt.week
    tmp['weekdayno'] = tmp.date.dt.weekday
    sub = tmp[tmp.year == year].copy()
    sub['value_norm'] = sub.value / (sub.value.max())
    subdf = sub[['date','value']].set_index("date")
    if ax==None:
        fig,ax = plt.subplots(figsize=(20,7))
    events2 = pd.Series(index=subdf.index, data=subdf.values.reshape(-1))
    return cp.yearplot(events2, year=year, ax=ax)
```

```python
# note -- preprocessingrequired --> 
# clean_data["SpeedLimitViolation"] = clean_data["SpeedLimitViolation"].astype("bool")

def dashboard_SpeedlimViolation_heatmap(subset, fillna=False,figsize=(10,10)):
    speedlimvio_hmap_df= pd.DataFrame(subset[subset["SpeedLimitViolation"]==True].groupby(["EntryGate","ExitGate"])["SpeedLimitViolation"].count())
    _df = speedlimvio_hmap_df.reset_index().pivot('EntryGate', 'ExitGate', 'SpeedLimitViolation')
    if fillna:
        _df.fillna(value=0, inplace=True)
    fig,ax = plt.subplots(figsize=figsize)
    ax.set_title("Speed Limit Violation - Heatmap across gates")
    sns.heatmap(_df, annot=False, fmt="f", linewidths=.5, cmap=sns.cm.rocket_r)
    plt.show()

def dashboard_SpeedlimViolationBar():
    out = widgets.Output()
    if str(clean_data["SpeedLimitViolation"].dtype) != "bool":
        clean_data["SpeedLimitViolation"] = clean_data["SpeedLimitViolation"].astype("bool")
    from_date = widgets.DatePicker(description='Start Date',disabled=False, value=clean_data.Date_dt.min())
    to_date = widgets.DatePicker(description='End Date',disabled=False,value=clean_data.Date_dt.max())

    FROM_DATE, TO_DATE =  from_date.value,  to_date.value
    
    def filter_records_based_on_start_and_end_time(clean_data,start,end):
        mask1 = clean_data.Date_dt >= pd.to_datetime(start)
        mask2 = clean_data.Date_dt <= pd.to_datetime(end)
        mask = mask1 & mask2
        return clean_data[mask]
    
    def on_change_from(change):
        nonlocal FROM_DATE
        if change['type'] == 'change' and change['name'] == 'value':
            FROM_DATE = change['new']
            render()
                
    def on_change_to(change):
        nonlocal TO_DATE
        if change['type'] == 'change' and change['name'] == 'value':
            TO_DATE = change['new']
            render()
            
    def render():
        with out:
            clear_output()
            if FROM_DATE and TO_DATE:
                subset = filter_records_based_on_start_and_end_time(clean_data,FROM_DATE, TO_DATE)
                res = subset[subset["SpeedLimitViolation"]==True].groupby(["EntryGate"])["SpeedLimitViolation"].count()
                fig,ax = plt.subplots(figsize=(20,8)) 
                plt.bar(res.index,res.values)
                plt.plot(res.index,res.values)
                _ = plt.title("Total Violation Across Gates")
                _ = plt.xticks( rotation='vertical')
                plt.show()
                #render heatmap
                dashboard_SpeedlimViolation_heatmap(subset)
                plt.show()
            else:
                print("INVALD DATE RANGE")
                
    from_date.observe(on_change_from)
    to_date.observe(on_change_to)
    
    line_1 = widgets.HBox([widgets.HTML(value="<h1>Violation of Cars Across Gates</h1>")])
    line_2 = widgets.HBox([from_date, to_date])
    W = widgets.VBox([line_1,line_2, out])
    
    display(W)
    render()
```

## Dwell time

```python

def dashboard_show_dwell_time_heatmap():
    # Any widget
    id_to_name,name_to_id = vechile_type_lookup()
    valid_car_type = [ id_to_name[v] for v in clean_data['car-type'].unique()] + ['ALL']
    valid_timecat = clean_data.time_cat.unique()
    
    vtype_dropdown = widgets.Dropdown(options=valid_car_type, value=valid_car_type[0], description='Vehicle Type:')
    time_menu = widgets.Dropdown(options=list(valid_timecat)+["ALL"], value=valid_timecat[0], description='Select Time slot:')
    
    def filter_records_based_on_vtype(subset, vtype):
        return subset[subset['car-type'] == vtype]
    def filter_records_based_on_time_cat(subset, time_cat):
        return subset[subset["time_cat"]==time_cat]

    out = widgets.Output()
    
    def render():
        with Timer("Filter based on Vtype"):
            if VTYPE == "ALL":
                subset = clean_data
            else:
                subset = filter_records_based_on_vtype(clean_data, name_to_id[VTYPE])
                
            if TIME_SLOT != "ALL":
                subset = filter_records_based_on_time_cat(subset, TIME_SLOT)
                
            print("No of records: ", subset.shape)
            show_heatmap_seconds_spent(subset, type_="sum")
            plt.show()
            show_heatmap_seconds_spent(subset, type_="mean")
            plt.show()

    VTYPE     = vtype_dropdown.value
    TIME_SLOT = valid_timecat[0]

    def on_change_vtype_dropdown(change):
        nonlocal VTYPE
        with out:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                print ("changed to %s" % change['new'])
                VTYPE = change['new']
                render()
    def on_time_menu_change(change):
        nonlocal TIME_SLOT
        with out:
            if change['type'] == 'change' and change['name'] == 'value':
                clear_output()
                newvalue = change['new']
                TIME_SLOT = newvalue
                render()
                
    #register a change listener
    vtype_dropdown.observe(on_change_vtype_dropdown)
    time_menu.observe(on_time_menu_change)

    # render the box
    line_1 = widgets.HBox([vtype_dropdown])
    W = widgets.VBox([line_1, out])
    display(W)
    
def show_heatmap_seconds_spent(subset, title="", fillna=False,
                               type_ = "mean",
                               figsize=(10,10), ax=None):
    res = subset.groupby(["EntryGate","ExitGate"])["SecondSpent"]
    if type_=="sum":
        title = "Dwell Time - Heatmap across gates (sum)"
        subset = pd.DataFrame(res.sum())
    else:
        title = "Dwell Time - Heatmap across gates (mean)"
        subset = pd.DataFrame(res.mean())
    _df = subset.reset_index().pivot('EntryGate', 'ExitGate', 'SecondSpent')
    if fillna:
        _df.fillna(value=0, inplace=True)
    if not ax: fig,ax = plt.subplots(figsize=figsize)
    ax.set_title(title)
    sns.heatmap(_df, annot=False, fmt="f", linewidths=.5, cmap=sns.cm.rocket_r, ax=ax)
    plt.show()
    
```

```python
import os
import sys

class suppress_stdout_stderr(object):
    '''
    A context manager for doing a "deep suppression" of stdout and stderr in
    Python, i.e. will suppress all print, even if the print originates in a
    compiled C/Fortran sub-function.
       This will not suppress raised exceptions, since exceptions are printed
    to stderr just before a script exits, and after the context manager has
    exited (at least, I think that is why it lets exceptions through).

    '''
    def __init__(self):
        self.null_fds = [os.open(os.devnull, os.O_RDWR) for x in range(2)]
        self.save_fds = (os.dup(1), os.dup(2))

    def __enter__(self):
        os.dup2(self.null_fds[0], 1)
        os.dup2(self.null_fds[1], 2)

    def __exit__(self, *_):
        os.dup2(self.save_fds[0], 1)
        os.dup2(self.save_fds[1], 2)
        os.close(self.null_fds[0])
        os.close(self.null_fds[1])
```

```python
def forecast_gate_volumes(subset,duration=10):
    gp = subset.groupby(['EntryGate'])
    models , forecasts ,rmses = [],[], []
    forecast_dict = {}
    
    for group in gp.groups.keys():
        tmp = gp.get_group(group)
        ##get data to fit
        tmp = tmp.sort_values(['Date_dt'])
        tmp = tmp.groupby("Date_dt")['Date_dt'].count()
        #tmp.plot(kind="line")
        tmp = pd.DataFrame(tmp)
        tmp.columns = ['y']
        tmp = tmp.reset_index()
        tmp.columns = ['ds','y']

        model = Prophet()
        model.fit(tmp)
        future = model.make_future_dataframe(periods=duration)
        forecast = model.predict(future)

        #fig1 = model.plot(forecast)
        forecast_orginal = forecast.iloc[:-duration][['ds','yhat','yhat_lower','yhat_upper']]
        forecast_recent = forecast.iloc[-duration:][['ds','yhat','yhat_lower','yhat_upper']]
        f_value = forecast_recent.yhat.sum()
        forecasts.append(forecasts)
        models.append(model)
        forecast_dict[group] = f_value

        #print metrics: mse
        assert len(tmp['y']) == len(forecast_orginal['yhat']), "Mismatch in y and y_hat lenght"
        rmse = mean_squared_error(tmp['y'],forecast_orginal['yhat'], squared=False)
        #print(rmse)
        rmses.append(rmse)
    
    return forecast_dict, models , forecasts ,rmses
```

```python
def Draw_GateVolume_Forecase_Graph(G, forecast_dict, custompos, figsize = (10,10), 
                           edge_scale=0.0012, node_scale=1.5):
    
    node_size = [ node_scale*forecast_dict[v] for v in G] 
           
    edge_width = [edge_scale * G[u][v]['weight'] for u, v in G.edges()] 
    
    fig,ax = plt.subplots(figsize=(8,8))
    nx.draw(G,custompos, node_color='r', ax=ax, with_labels = True,alpha = .8, width = edge_width,
            edge_color='gray',
            node_size=node_size)
    
    plt.axis('equal')
    plt.tight_layout();
    return fig
```
